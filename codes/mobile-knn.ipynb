```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
import pickle
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
with open('X_train.pickle', 'rb') as handle:
    X_train = pickle.load(handle)
    
with open('X_test.pickle', 'rb') as handle:
    X_test = pickle.load(handle)
    
with open('y_train.pickle', 'rb') as handle:
    y_train = pickle.load(handle)

with open('y_test.pickle', 'rb') as handle:
    y_test = pickle.load(handle)
```

```python

parameters = {
    'n_neighbors':range(0, 20),
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
     'weights':['distance' , 'uniform'],
    'p':[1, 2],
}

KNN = KNeighborsClassifier()
clf = GridSearchCV(KNN, parameters ,n_jobs = -1 ,scoring = 'accuracy' , cv = 5 , verbose=4)
```

```python
clf.fit(X_train,y_train)
```

```python
p = clf.predict(X_test)
e = y_test
print(metrics.classification_report(e,p))
```

```python
df_test = pd.read_csv('../datasets/test.csv')
df_test.drop(['id'] ,axis=1,inplace =True)
result = clf.predict(df_test.to_numpy())
df_test['result'] = pd.Series(result)
df_test.to_csv('../datasets/mobile-knn-grid-search.csv')
df_test
```

```python
parameters = {
    'n_neighbors':range(0, 100),
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
     'weights':['distance' , 'uniform'],
    'p':[1, 2],
    'leaf_size' : range(10 , 120)
}

KNN = KNeighborsClassifier()
clf = RandomizedSearchCV(KNN, parameters ,n_jobs = -1 ,scoring = 'f1_macro' , cv = 6 , verbose=4 , random_state=0 ,n_iter =100)
```

```python
clf.fit(X_train,y_train)
```

```python
p = clf.predict(X_test)
e = y_test
print(metrics.classification_report(e,p))
```

```python
df_test = pd.read_csv('../datasets/test.csv')
df_test.drop(['id'] ,axis=1,inplace =True)
result = clf.predict(df_test.to_numpy())
df_test['result'] = pd.Series(result)
df_test.to_csv('../datasets/mobile-knn-rand-grid-search.csv')
df_test
```
