```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense , Conv1D , MaxPool1D , Flatten
from tensorflow.keras.utils import to_categorical
import tensorflow as tf
from tensorflow.keras.datasets import mnist
```

```python
df = pd.read_csv('../datasets/train.csv')
df
```

```python
df.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data = scaler.fit_transform(df.iloc[: , :-1])
data
```

```python
X = data
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
np.shape(X_train)
```

```python
X_train = np.expand_dims(X_train, axis=2)
X_test = np.expand_dims(X_test, axis=2)
X_train
```

```python
model = Sequential([
    Conv1D(32, 3, activation='relu' , input_shape=(20,1)),
    Conv1D(64, 3, activation='relu'),
    Conv1D(128, 3, activation='relu'),
    Dense(256, activation='relu'),
    Dense(128, activation='relu'),
    Flatten(),
    Dense(4, activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile('adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
y_train_ohe = to_categorical(y_train)
y_test_ohe = to_categorical(y_test)
```

```python
mcp = model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath='../models/cnn.h5',
    save_weights_only=True,
    monitor='val_accuracy',
    mode='max',
    save_best_only=True)
```

```python
model.fit(X_train, y_train_ohe, epochs=100, validation_data=(X_test, y_test_ohe), batch_size=32 , callbacks=[mcp])
```

```python
model.load_weights('../models/cnn.h5')
```

```python
df_test = pd.read_csv('../datasets/test.csv')
df_test.drop(['id'] ,axis=1,inplace =True)
```

```python
X_test = scaler.transform(df_test)
```

```python
X_test = np.expand_dims(X_test, axis=2)
```

```python
y_pred = model.predict(np.array(X_test))
y_pred = [np.argmax(x) for x in y_pred]
df_test['result'] = pd.Series(y_pred)
df_test.to_csv('../datasets/mobile-cnn.csv')
df_test
```
