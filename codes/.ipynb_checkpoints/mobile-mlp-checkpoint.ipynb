```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import pylab as pl
import seaborn as sns
from fitter import Fitter, get_common_distributions, get_distributions
```

```python
df = pd.read_csv('train.csv')
pd.set_option('display.max_columns', None)
df
```

```python
df.isnull().sum()
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
X = mms.fit_transform(X)
```

```python
X
```

```python
y
```

```python
y = pd.get_dummies(y)
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=3)
```

```python
y_train.shape
```

```python
model = tf.keras.models.Sequential(
[
    tf.keras.layers.Dense(20 , activation='relu' , input_shape=(20,)),
    tf.keras.layers.Dense(100 , activation='relu'),
    tf.keras.layers.Dense(20 , activation='relu'),
    tf.keras.layers.Dense(4 , activation='softmax')    
])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(0.001) , loss = tf.keras.losses.CategoricalCrossentropy() , metrics=[tf.keras.metrics.CategoricalAccuracy()])
```

```python
model.fit(X_train, y_train, epochs=100, validation_data=(X_test, y_test), batch_size=24)
```
