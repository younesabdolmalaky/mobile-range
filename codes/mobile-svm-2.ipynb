```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
```

```python
with open('X_train.pickle', 'rb') as handle:
    X_train = pickle.load(handle)
    
with open('X_test.pickle', 'rb') as handle:
    X_test = pickle.load(handle)
    
with open('y_train.pickle', 'rb') as handle:
    y_train = pickle.load(handle)

with open('y_test.pickle', 'rb') as handle:
    y_test = pickle.load(handle)
```

```python
from sklearn.svm import SVC
from sklearn import metrics
svm_classifier = SVC(C=1 ,kernel = 'linear',probability = True)
svm_classifier.fit(X_train, y_train)
p = svm_classifier.predict(X_test)
e = y_test
print(metrics.classification_report(e, p))
```

```python
df_test = pd.read_csv('../datasets/test.csv')
df_test.drop(['id'] ,axis=1,inplace =True)
result = svm_classifier.predict(df_test.to_numpy())
df_test['result'] = pd.Series(result)
df_test.to_csv('../datasets/mobile-svm.csv')
df_test
```

```python
from sklearn.model_selection import RandomizedSearchCV

parameters = {
    'C':np.arange(0, 50 , 0.5),
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid' ],
    'decision_function_shape':['ovo', 'ovr'],
    'degree':range(1,100,2),
    'gamma':['scale', 'auto'],
    'coef0':np.arange(0,20 , 0.2)
}

clf = RandomizedSearchCV(SVC(), parameters ,n_jobs = -1 ,scoring = 'f1_macro' , cv = 6 , verbose=4 , random_state=0 ,n_iter =10)
```

```python
clf.fit(X_train,y_train)
```

```python
p = clf.predict(X_test)
e = y_test
print(metrics.classification_report(e,p))
```

```python
df_test = pd.read_csv('../datasets/test.csv')
df_test.drop(['id'] ,axis=1,inplace =True)
result = clf.predict(df_test.to_numpy())
df_test['result'] = pd.Series(result)
df_test.to_csv('../datasets/mobile-svm-rand-grid-search.csv')
df_test
```
