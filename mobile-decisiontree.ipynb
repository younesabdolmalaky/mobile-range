```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
df = pd.read_csv('train.csv')
df
```

```python
X = df.iloc[ : , df.columns != 'price_range'].to_numpy()
Y = df['price_range'].to_numpy()
```

```python
from sklearn.tree import DecisionTreeClassifier
des_tree_classifier = DecisionTreeClassifier(criterion= 'entropy')
```

```python
des_tree_classifier.fit(X[:1500],Y[:1500])
```

```python
from sklearn import metrics
```

```python
p = des_tree_classifier.predict(X[1500:])
e = Y[1500:]
print(metrics.classification_report(e,p))
```

```python
df_test = pd.read_csv('test.csv')
df_test.drop(['id'] ,axis=1,inplace =True)
df_test
```

```python
probability = des_tree_classifier.predict_proba(df_test.to_numpy())
result = des_tree_classifier.predict(df_test.to_numpy())
temp_df = pd.DataFrame(probability)
probability = temp_df.max(axis = 1).to_numpy()
```

```python
new_data_frame = df_test
new_data_frame['result'] = pd.Series(result)
new_data_frame['probability'] = pd.Series(probability)
new_data_frame
```

```python
new_data_frame.to_csv('mobile-decisiontree.csv')
```
