```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
df
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ADA = AdaBoostClassifier()
X = df.iloc[ : , df.columns!='price_range']
Y = df['price_range']
```

```python
ADA.fit(X[:1500] , Y[:1500])
```

```python
from sklearn import metrics
```

```python
p = ADA.predict(X[1500:])
e = Y[1500:]
print(metrics.classification_report(e,p))
```

```python
df_test = pd.read_csv('test.csv')
df_test.drop(['id'] ,axis=1,inplace =True)
df_test
```

```python
probability = ADA.predict_proba(df_test.to_numpy())
result = ADA.predict(df_test.to_numpy())
temp_df = pd.DataFrame(probability)
probability = temp_df.max(axis = 1).to_numpy()
```

```python
new_data_frame = df_test
new_data_frame['result'] = pd.Series(result)
new_data_frame['probability'] = pd.Series(probability)
new_data_frame
```

```python
new_data_frame.to_csv('mobile-adaboost.csv')
```
